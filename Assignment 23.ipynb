## Q3. Given a Binary Tree, print the Right view of it.

The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input:**

         1

      /     \

   2         3

/   \       /  \

4     5   6    7

             \

               8

**Output**:

Right view of the tree is 1 3 7 8

**Input:**

         1

       /

    8

  /

7

**Output**:

Right view of the tree is 1 8 7


```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def print_right_view(root):
    if root is None:
        return

    queue = [(root, 0)]
    right_view = []

    while queue:
        node, level = queue.pop(0)

        # If this is the last node at the current level
        if level == len(right_view):
            right_view.append(node.val)

        # Traverse the right subtree before the left subtree
        if node.right:
            queue.append((node.right, level + 1))
        if node.left:
            queue.append((node.left, level + 1))

    print("Right view of the tree is:", *right_view)

# Construct the binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)
root.right.right.right = TreeNode(8)

# Print the right view of the tree
print_right_view(root)
```

## Q4. Given a Binary Tree, The task is to print the **bottom view** from left to right. A node **x** is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input:**

             20

           /     \

        8         22

    /      \         \

5         3        25

        /    \

   10       14

**Output:** 5, 10, 3, 14, 25.

**Input:**

             20

           /     \

        8         22

    /      \      /   \

 5         3    4     25

         /    \

     10       14

**Output:**

5 10 4 14 25.

**Explanation:**

If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.

**3 and 4** are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

```python
from collections import deque

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def print_bottom_view(root):
    if root is None:
        return

    # Dictionary to store nodes at each horizontal distance
    horizontal_distance_map = {}

    # Queue for level-order traversal
    queue = deque([(root, 0)])

    while queue:
        node, horizontal_distance = queue.popleft()

        # Update the node at the current horizontal distance
        horizontal_distance_map[horizontal_distance] = node.val

        # Enqueue the left child
        if node.left:
            queue.append((node.left, horizontal_distance - 1))

        # Enqueue the right child
        if node.right:
            queue.append((node.right, horizontal_distance + 1))

    # Print the nodes from the dictionary
    for key in sorted(horizontal_distance_map.keys()):
        print(horizontal_distance_map[key], end=" ")
    print()

# Construct the binary tree
root = TreeNode(20)
root.left = TreeNode(8)
root.right = TreeNode(22)
root.left.left = TreeNode(5)
root.left.right = TreeNode(3)
root.right.right = TreeNode(25)
root.left.right.left = TreeNode(10)
root.left.right.right = TreeNode(14)

# Print the bottom view of the tree
print_bottom_view(root)
```
